```python
%matplotlib notebook
import numpy as np
import ijroi
import glob
from scipy.interpolate import Rbf
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import ipyvolume.pylab as p3
x, y, z = [], [], []
rois = glob.glob("C:/Users/kiritani/Documents/data/cells/TK577/Anatomy/SurfaceRoi/*.roi")
for f in rois:
    with open(f, 'rb') as fobj:
        roi = ijroi.read_roi(fobj)
        x.append(roi.points[0][0])
        y.append(roi.points[0][1])
        z.append(roi.position)
        
smoothed = Rbf(x, y, z, smooth=1)
vfunc = np.vectorize(smoothed)
X = np.arange(100, 1000)
Y = np.arange(100, 1000)
X, Y = np.meshgrid(X, Y)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X*0.6, Y*0.6, vfunc(X, Y)*2.95,
                       linewidth=0, antialiased=False)

rois_neurons = glob.glob("C:/Users/kiritani/Documents/data/cells/TK577/Anatomy/RoiSet/*.roi")

x_neurons, y_neurons, z_neurons = [], [], []
for f in rois_neurons:
    with open(f, 'rb') as fobj:
        roi = ijroi.read_roi(fobj)
        x_neurons.append(roi.points[0][0] * 0.6)
        y_neurons.append(roi.points[0][1] * 0.6)
        z_neurons.append(roi.position * 2.95)
ax.scatter(x_neurons, y_neurons, z_neurons)
```

```python
import ipyvolume.pylab as p3
fig = p3.figure()
sc1 = p3.scatter(np.array(x_neurons), np.array(y_neurons), np.array(z_neurons), marker='sphere', size=1)
#sc2 = p3.scatter(np.array(x_neurons), np.array(y_neurons), np.array(z_neurons)/5, color='blue', marker='box',size=1)
pia = p3.scatter(X.flatten()[::5] * 0.6, Y.flatten()[::5] * 0.6, vfunc(X.flatten(), Y.flatten())[::5]*2.95, color='blue', marker='box', size=0.5)
p3.xlim(-200, 800)
p3.ylim(-200, 800)
p3.zlim(0, 1000)
p3.show()
```
